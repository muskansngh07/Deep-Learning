```python
from math import e as e
```

```python
def activation(x):
    return 1/(1+e**(-x))
```

```python
def forward_pass(x,w_hid,b_hid,w_out,b_out):
    hidden=[]
    for i in range(len(b_hid)):
        result=sum(x[j]*w_hid[i][j] for j in range(len(x)))+b_hid[i]
        hidden.append(activation(result))

    output=[]
    for i in range(len(b_out)):
        result=sum(hidden[j]*w_out[i][j] for j in range(len(hidden)))+b_out[i]
        output.append(activation(result))

    return hidden,output
```

```python
# Enter the number of neurons in the input layer
n=int(input("Enter the number of input neurons: "))

# Enter the value of neurons in the input layer
x=list(map(float,input("Enter the value of input neurons separated by a space: ").split()))

# Enter the number of hidden layer neurons
h=int(input("Enter the number of neurons in the hidden layer: "))

w_hid=[]
b_hid=[]
for i in range(h):
    w=list(map(float,input("Enter the weights of associated with hidden neuron "+str(i+1)+": ").split()))
    b=float(input("Enter the bias associated with hidden neuron "+str(i+1)+": "))
    w_hid.append(w)
    b_hid.append(b)

# Enter the number of neurons in the output layer
o=int(input("Enter the number of neurons in the output layer: "))

w_out=[]
b_out=[]
for i in range(o):
    w=list(map(float,input("Enter the weights of associated with hidden neuron "+str(i+1)+": ").split()))
    b=float(input("Enter the bias associated with hidden neuron "+str(i+1)+": "))
    w_out.append(w)
    b_out.append(b)
```

```python
print(forward_pass(x,w_hid,b_hid,w_out,b_out))
```
